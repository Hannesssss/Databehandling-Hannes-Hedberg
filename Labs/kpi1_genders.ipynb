# KPI1 - Vaccination Rate for men and women

The vaccination rate differs between men and women, this KPI is to investigate the differences between the two genders. And plot of some graphs, charts or diagrams to display the difference between the genders.

## 0. Importing tools

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly_express as px
```

----

## 1. Reading the Data

Using Pandas we read the excel file, and open the sheet "vaccinerade kön"

```python
genders = pd.read_excel('../Data/Folkhalsomyndigheten_Covid19_Vaccine.xlsx', sheet_name='Vaccinerade kön')
genders
```

To have a better idea of what we are working with we print out some information about the genders sheet.

```python
display(genders.info(),
        genders.describe(),
        genders.head(10))
```

----

## 2. Analyzing the vaccination rate

```python
num_vaccinated_1_dose = genders.loc[:0, "Antal vaccinerade"].sum()
num_vaccinated_2_doses = genders.loc[1:1, "Antal vaccinerade"].sum()
print(f"Total who took atleast 1 dose: {num_vaccinated_1_dose}, Total who took atleast 2 doses: {num_vaccinated_2_doses}")
```

```python
num_men_vaccinated_1_dose = genders.loc[2:2, "Antal vaccinerade"].sum()
num_women_vaccinated_1_dose = genders.loc[4:4, "Antal vaccinerade"].sum()
print(f"Men who took atleast 1 dose: {num_men_vaccinated_1_dose}, Women who took atleast 1 dose: {num_women_vaccinated_1_dose}")
```

```python
num_men_vaccinated_2_doses = genders.loc[3:3, "Antal vaccinerade"].sum()
num_women_vaccinated_2_doses = genders.loc[5:5, "Antal vaccinerade"].sum()
print(f"Men who took atleast 2 doses: {num_men_vaccinated_2_doses}, Women who took atleast 2 doses: {num_women_vaccinated_2_doses}")
```

----

## 3. Piecharts

Plotting out a few piecharts to display the difference in vaccination rate inbetween the genders.

### 3.1 Pie chart - 1st Dose

```python
pie1 = np.array([num_men_vaccinated_1_dose, num_women_vaccinated_1_dose])
mylabels = ["Men", "Women"]
myexplode = [0.2, 0]
mycolors = ["blue", "hotpink"]

fig, ax = plt.subplots()

plt.pie(pie1, labels = mylabels, explode = myexplode, shadow = True, colors = mycolors)
plt.title("First Dose: Gender Difference", y=1.08)
plt.legend(title = "Genders")
plt.show()
plt.savefig("Visualiseringar/KPI1_Gender_vaccination_equality.png", dpi=100) 
```

```python
pie1 = np.array([num_men_vaccinated_1_dose, num_women_vaccinated_1_dose])
mylabels = ["Men", "Women"]
myexplode = [0.2, 0]
mycolors = ["blue", "hotpink"]

fig, ax = plt.subplots()

ax.pie(pie1, labels = mylabels, explode = myexplode, shadow = True, colors = mycolors)
ax.set_title("First Dose: Gender Difference", y=1.08)
ax.legend(title = "Genders")
fig.show()
plt.savefig("Visualiseringar/KPI1_Gender_vaccination_equality.png", dpi=100)
```

```python
pie2 = np.array([num_vaccinated_1_dose, num_men_vaccinated_1_dose, num_men_vaccinated_1_dose])
mylabels = ["Total", "Men", "Women"]
myexplode = [0.1, 0.1, 0]
mycolors = ["green", "blue", "hotpink"]

plt.pie(pie2, labels = mylabels, explode = myexplode, shadow = True, colors = mycolors)
plt.title("First Dose Total", y=1.08)
plt.legend(title = "Genders")
plt.show() 
plt.savefig("Visualiseringar/KPI1_Gender_vaccination_equality_and_total.png", dpi=100)
```

### 3.2 Pie chart - 2nd Dose

```python
pie3 = np.array([num_men_vaccinated_2_doses, num_women_vaccinated_2_doses])
mylabels = ["Men", "Women"]
myexplode = [0.2, 0]
mycolors = ["blue", "hotpink"]

plt.pie(pie3, labels = mylabels, explode = myexplode, shadow = True, colors = mycolors)
plt.title("Second Dose: Gender Difference", y=1.08)
plt.legend(title = "Genders")
plt.show()
plt.savefig("Visualiseringar/KPI1_Gender_vaccination_equality_2nd_dose.png", dpi=100) 
```

```python
pie4 = np.array([num_vaccinated_2_doses, num_women_vaccinated_2_doses, num_men_vaccinated_2_doses])
mylabels = ["Total", "Men", "Women"]
myexplode = [0.1, 0.1, 0]
mycolors = ["green", "blue", "hotpink"]
fig, ax = plt.subplots()

plt.pie(pie4, labels = mylabels, explode = myexplode, shadow = True, colors = mycolors)
plt.title("Second Dose: Total", y=1.08)
plt.legend(title = "Genders")
plt.show() 
plt.savefig("Visualiseringar/KPI1_Gender_vaccination_equality_and_total_2nd_dose.png", dpi=100)
```

### 3.3 Detailed Pie chart

this was taken from reference1, in order to display numbers in the pie charts

```python
labels = 'Total', 'Men', 'Women'
sizes = np.array([num_vaccinated_2_doses, num_women_vaccinated_2_doses, num_men_vaccinated_2_doses])
colors = ['green', 'blue', 'hotpink']
myexplode = [0.2, 0, 0]

def absolute_value(val):
    a  = np.round(val/100.*sizes.sum(), 0)
    return a

plt.pie(sizes, labels=labels, colors=colors,
        autopct=absolute_value, shadow=True, explode = myexplode)

plt.axis('equal')
plt.savefig("Visualiseringar/KPI1_Gender_vaccination_equality_and_total_2nd_dose_detailed.png", dpi=100)
plt.show()
```

Since the amount of difference between genders is minimal, both pie charts are almost equal. 

```python
labels = 'Total', 'Men', 'Women'
sizes = np.array([num_vaccinated_1_dose, num_women_vaccinated_1_dose, num_men_vaccinated_1_dose])
colors = ['green', 'blue', 'hotpink']
myexplode = [0.2, 0, 0]

def absolute_value(val):
    a  = np.round(val/100.*sizes.sum(), 0)
    return a

plt.pie(sizes, labels=labels, colors=colors,
        autopct=absolute_value, shadow=True, explode = myexplode)

plt.axis('equal')
plt.savefig("Visualiseringar/KPI1_Gender_vaccination_equality_and_total_detailed.png", dpi=100)
plt.show()
```
