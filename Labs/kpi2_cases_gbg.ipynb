## 0. Importing Tools

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

---

## 1. Reading the Data

```python
c19_num_cases = pd.read_excel('../Data/Folkhalsomyndigheten_Covid19.xlsx', sheet_name="Antal per dag region")   # Reads excel, and checks the "Antal per dag region" column
c19_num_cases.head(5)
```

---

## 2. Analyzing the Data

In this analysis we are looking at "antal per dag region" 

```python
display(c19_num_cases.info(),                   # just like in previous tasks we use display to check multiple blocks of info at the same time
        c19_num_cases.describe(),
        c19_num_cases.head(10))
```

---

## 3. Covid-19 Cases in Västra Götaland

```python
c19_num_cases = c19_num_cases[["Statistikdatum", "Västra_Götaland"]]        # Opens Statistikdatum, and Västra_Götaland columns, as displayed below
c19_num_cases
```

---

## 4. Plotting cases in Västra Götaland

```python
fig = sns.lineplot(                                                     # trying out seaborn to print a lineplot
    data=c19_num_cases,                                                 # sets the data to be taken from our dataframe
    x="Statistikdatum",                                                 # locates "Statistikdatum", and assigned it to the x label
    y="Västra_Götaland",                                                # locates "Västra_Götaland", and assigns it to the y label 
)

fig.set(
    title="Number of daily Covid-19 cases in Västra Götaland",          # displays title above the line 
    xlabel="",                                                          # instead of typing individual values, we pull all "Statistikdatum" into x & y
    ylabel=""
    
)
plt.xticks(fontsize=7, rotation=90)                                    # Suggested by Anna-Maria, to prevent the dates from colliding
fig = fig.get_figure() 
fig.savefig("Visualiseringar/KPI2_Daily_Cases_Vastra_Gotaland.png")     # exports the image                                           
```
