## 1. Reading the Data

## 2. Covid-19 Cases in Västra Götaland

```python
c19_num_cases = c19_num_cases[['Statistikdatum', 'Västra_Götaland']]
c19_num_cases
```

## 3. Plotting cases in Västra Götaland

### 3.1 Lineplot of daily cases in Västra Götaland

```python
fig = sns.lineplot(
    data=c19_num_cases,
    x="Statistikdatum",
    y="Västra_Götaland",
)

fig.set(
    title='Number of daily cases in Västra Götaland',
    xlabel='',
    ylabel=''
)

fig = fig.get_figure()
```

### 3.2 Total cases in Sweden

How many of the cases in all of Sweden were in Västra Götaland? 
(Totalt antal per region -> totalt_antal_fall)

```python
c19_num_total_cases = pd.read_excel('../Data/Folkhalsomyndigheten_Covid19.xlsx', sheet_name='Totalt antal per region')
c19_num_total_cases.head()
```

#### 3.2.1 Sum of all cases in Sweden

```python
num_sweden_cases = c19_num_total_cases["Totalt_antal_fall"].sum()
num_sweden_cases
```

#### 3.2.2 Sum of all cases in Västra Götaland

```python
locate_vastra_gotaland_cases = c19_num_total_cases.loc[18:18, "Region"].sum()
locate_vastra_gotaland_cases
```

```python
num_vastra_gotaland_cases = c19_num_total_cases.loc[18:18, "Totalt_antal_fall"].sum()
num_vastra_gotaland_cases
```

### 3.3 Pie chart

```python
labels = 'Västra Götaland', 'Sweden'
sizes = np.array([num_vastra_gotaland_cases, num_sweden_cases])
colors = ['cyan', 'yellow']

def absolute_value(val):
    a  = np.round(val/100.*sizes.sum(), 0)
    return a

plt.pie(sizes, labels=labels, colors=colors,
        autopct=absolute_value, shadow=True)

plt.axis('equal')
plt.show()
```
